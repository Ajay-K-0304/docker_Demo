```python
#Partitioning and repartitioning
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, desc, sum, avg
# Initialize
spark = SparkSession.builder.appName("SparkCompleteNotes").getOrCreate()
# Create Base DataFrame
data = [
    (1, "Alice", "Engineering", 75000, 25),
    (2, "Bob", "Marketing", 60000, 30),
    (3, "Charlie", "Engineering", 80000, 35),
    (4, "David", "Sales", 65000, 28),
    (5, "Eve", "Marketing", 70000, 32),
    (6,"Kat","Sales",50000,34),
    (7,"Ajay","Engineering",90000,23),
    (8,"Ashiv","Sales",85000,24),
    (9,"Arpita","Marketing",48000,21),
    (10,"Arpit","Engineering",70000,21),
    (11,"Anirav","Marketing",50000,23),
    (12,"Aditya","Sales",45000,21),
    (13,"Abhishree","Engineering",70000,21),
    (14,"Aarohi","Sales",70000,23),
    (15,"Hiya","Marketing",75000,21),
    (16,"Diya","Marketing",60000,23),
    (17,"Maitri","Sales",100000,23),
    (18,"Aish","Engineering",87000,22),
    (19,"Morwal","Sales",78000,24),
    (20,"Shreyanshi",90000,21)
]
columns = ["Id", "Name", "Department", "Salary", "Age"]
df = spark.createDataFrame(data, columns);
```

```python
numData = spark.range(0,1000000).withColumn("value",col("id")%1000)
```

```python
numData.take(10)
```

```python
print("Before partition:",numData.rdd.getNumPartitions())
```

```python
numData.write.mode("overwrite").csv("output/beforeEmp",header=True)
```

```python
numData_repartition=df.repartition(20)
print("After partition:",numData_repartition.rdd.getNumPartitions())
```

```python
numData_repartition.write.mode("overwrite").csv("output/afterEmp",header=True)
```

```python
numData_coal=numData_repartition.coalesce(2)
print("After coalesced:",numData_coal.rdd.getNumPartitions())
```
